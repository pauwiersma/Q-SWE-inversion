```python
#Script to check the netcdf mean timeseries
import xarray as xr
import glob
import matplotlib.pyplot as plt

def dailymeans(ERAncfile):
    if 'OBS6' in ERAncfile:
        ds = xr.open_dataset(ERAncfile)
        ds_reduced = ds.sel(lon=slice(0,10),lat=slice(40,50))
        ds_mean = ds_reduced.mean(dim=('lat','lon'))
        var = list(ds_mean.variables)[0]
        print(ds[var].units)
        return ds_mean[var]
    else:
        ds = xr.open_dataset(ERAncfile)
        ds_reduced = ds.sel(longitude=slice(0,10),latitude=slice(50,40))
        ds_mean = ds_reduced.mean(dim=('latitude','longitude'))
        var = list(ds_mean.variables)[1]
        print(ds[var].units)
        return ds_mean[var]
```

```python
for file in files:
    if 'orog' in file:
        continue
    series = dailymeans(file)
    print(series.size)
    series.plot()
    print(series.time)
    plt.show()
```
